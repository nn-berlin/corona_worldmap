```python
import pandas as pd
import re
```

```python
pd.options.display.max_rows = 250
```

```python
df = pd.read_csv('./data/countries_raw.csv', index_col=False)
```

```python
df.columns
```

```python
df.drop(['Staat oder Gebiet', 'Langform des Staatsnamens', 'Hauptstadt','Flächein km²', 'Einw.pro km²', 'Einw.pro km²', 'Flag­ge', 'TLD', 'Lokaler Name', ], axis=1, inplace=True)
```

```python
df.columns
```

```python
df = df.rename(columns={"Englischer Name": "country", "ISO-3-Kürzel": "ISO3",
         'ISO-2-Kürzel': 'ISO2', 'Einwohner2017': 'population2017'})
```

```python
df = df.loc[2:206]
```

```python
# extracting numbers
df['population2017'] = df['population2017'].apply(
    lambda x: int(
        re.findall(
            r'\d+', str(x).replace('.', '')
        )[0]
    )
)
```

```python
# set iso3 Kosovo
df.loc[df[df.country == 'Kosovo'].index,'ISO3'] = 'XKX'
```

```python
# set iso2 namibia
df.loc[df[df.country == 'Namibia'].index,'ISO2'] = 'NA'
```

```python
df[df.ISO3.isna()]
```

```python
df[df.ISO2.isna()]
```

```python
# drop countries not in covid data
dropindex = df[df.ISO2.isna()].index
df.drop(index=dropindex, inplace=True)
df.reset_index(inplace=True, drop=True)
```

```python
# cut ISO2
df['ISO2'] = df['ISO2'].apply(lambda x: str(x)[:2])
```

```python
# removing alternative names 
df.country = df.country.apply(lambda x: x.split('oder')[0].replace("'", ''))
df.country = df.country.apply(lambda x: x.split(' (')[0].strip())
```

```python
df.to_csv('./data/countries.csv', index=False)
```
